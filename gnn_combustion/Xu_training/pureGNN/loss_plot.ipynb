```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = os.path.join('ten_tests', '6', 'train.txt')

if os.path.exists(file_path):
    data = np.loadtxt(file_path, skiprows=1, max_rows=21, delimiter='|', usecols=(0, 1, 2, 3),
                      converters={
                          0: lambda s: float(s.decode().replace('Epoch:', '')),
                          1: lambda s: float(s.decode().replace('Total loss:', '')),
                          2: lambda s: float(s.decode().replace('Bond loss:', '')),
                          3: lambda s: float(s.decode().replace('Existence loss:', ''))
                      })
    epochs, total, bond, existence = data.T

    fig, ax = plt.subplots(figsize=(9, 6))

    ax.plot(epochs, total, color='dodgerblue', lw=3.0, alpha=1.0, label='Total Loss')
    ax.plot(epochs, bond, color='orange', lw=3.0, alpha=1.0, label='Bond Loss')
    ax.plot(epochs, existence, color='green', lw=3.0, alpha=1.0, label='Existence Loss')

    # Styling
    ax.set_xticks([0, 50, 100, 150, 200])
    #ax.set_yscale('log')
    ax.set_xlabel('Epoch', fontsize=18, fontweight='bold')
    ax.set_ylabel('Loss Value', fontsize=18, fontweight='bold')
    ax.tick_params(axis='both', labelsize=16)
    ax.legend(fontsize=16, frameon=False)
    
    ax.set_title('Training Loss for One Single Run', fontsize=16, fontweight='bold')
    #ax.grid(True, which="both", ls="-", alpha=0.2)
    plt.tight_layout()
    plt.show()
```

```python
def pauling_bo(R, Re, b=0.357):
    return np.exp((Re - R) / b)

# Equilibrium bond lengths (Re) in Angstroms
bond_params = {
    'H-H': {'Re': 0.74, 'color': 'blue', 'linestyle': '-'},
    'O-H': {'Re': 0.96, 'color': 'green', 'linestyle': '--'},
    'O-O': {'Re': 1.48, 'color': 'red', 'linestyle': ':'}
}

# Distance range (Angstroms)
R = np.linspace(0.5, 3.5, 300)

fig, ax = plt.subplots(figsize=(10, 5.5))

for bond, params in bond_params.items():
    bo = pauling_bo(R, params['Re'])
    plt.plot(R, bo, label=bond, color=params['color'], 
             linestyle=params['linestyle'], linewidth=5)

# Reference lines for BO=1 (single bond) and BO=0
#plt.axhline(y=1.0, color='gray', linestyle='-', alpha=0.3)
#plt.axhline(y=0.0, color='gray', linestyle='-', alpha=0.3)

plt.xlabel('Interatomic Distance ($\AA$)', fontsize=25)
plt.ylabel('Pauling Bond Order', fontsize=25)
# Tick Label Font Size
ax.tick_params(axis='both', labelsize=25)
#plt.title("Pauling's Bond Order Estimation", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=25, frameon=False)
plt.ylim(-0.1, 2.5) 
plt.xlim(0.5, 3.5)

plt.savefig('pauling_bond_order.png')
plt.savefig('pauling_bond_order.svg')
```
